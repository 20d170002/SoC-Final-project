```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,5,11)
y = x**2
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('TITLE')
```

```python
#multiple plots on the same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('TITLE')
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('larger plot')

axes2.plot(y,x)
axes2.set_title('smaller plot')
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.1,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout()
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label ='x squared')
ax.plot(x,x**3, label = 'x cubed')

ax.legend(loc=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', linewidth = 1, linestyle= '-', marker ='o', markersize=20, markerfacecolor='yellow',
        markeredgewidth=3, markeredgecolor='green') #alpha is the transparency factor #lw works instead of linewidth #ls works as linestyle
```
