```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/content/master.csv')
df
```

```python
df.describe()
```

Mode for each variable

```python
 
df.mode(axis=0,numeric_only =True, dropna =True)
```

Median

```python
df.median(axis =0)
```

Skew

```python
print(df.skew(axis = 0, skipna = True))
```

Interquartile range

```python
my_list= ['year', 'suicides_no','population','suicides/100k pop','HDI for year','gdp_per_capita ($)']
for col in my_list:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  print(IQR)
```

```python
my_list= ['year', 'suicides_no','population','suicides/100k pop','HDI for year','gdp_per_capita ($)']
for col in my_list:
  df[col].hist()
  plt.show()
plt.tight_layout()
```

Too few bins 

```python
my_list= ['year', 'suicides_no','population','suicides/100k pop','HDI for year','gdp_per_capita ($)']
for col in my_list:
  df[col].hist(bins = 5)
  plt.show()
plt.tight_layout()
```

Too many bins

```python
my_list= ['year', 'suicides_no','population','suicides/100k pop','HDI for year','gdp_per_capita ($)']
for col in my_list:
  df[col].hist(bins = 100)
  plt.show()
plt.tight_layout()
```

Good number of bins

```python
my_list= ['year', 'suicides_no','population','suicides/100k pop','HDI for year','gdp_per_capita ($)']
for col in my_list:
  df[col].hist(bins = 25)
  plt.show()
plt.tight_layout()
```

Box-and- whiskers plot

```python
my_list= ['year', 'suicides_no','population','suicides/100k pop','HDI for year','gdp_per_capita ($)']
for col in my_list:
  df[col].plot(kind='box')
  plt.show()
  plt.tight_layout()
```
