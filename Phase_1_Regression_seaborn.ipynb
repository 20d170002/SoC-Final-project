```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips,hue ='sex',markers = ['o','v'], scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, col = 'sex', row = 'time',hue='sex')
```

```python
sns.set_style('white')
sns.countplot(x='sex', data =tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data =tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data =tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data =tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data =tips)
sns.despine()
```
